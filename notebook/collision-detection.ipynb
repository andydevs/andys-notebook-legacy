```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
from numpy.random import default_rng

# Ball position
rng = default_rng()
x = rng.random((20, 2))
v = rng.random((20, 2))

# Plot
plt.quiver(x[:,0], x[:,1], v[:,0], v[:,1]);
plt.xticks([]);
plt.yticks([]);
plt.show();
```
